cette variable nous indique la taille du malle :

```python
#changer le nom de la colonne pour faciliter le travail
data_cleaned_3=data_cleaned_3.rename(columns={'Trunk or Cargo Capacity': 'trunk_capacity'})
data_cleaned_3['trunk_capacity']
```

```python
trunk=[]
import math
for c in data_cleaned_3.trunk_capacity :
    if str(c).lower()=="nan":
         trunk.append(100.0)
    else :     
        c1=c[:-7]
        if c1[-1]==' ' :
            trunk.append(float(c1[:-1]))
        else :
            trunk.append(float(c1))        
    
data_cleaned_3['trunk_cap']=trunk

data_cleaned_3
```

Je vais transformer la variable trunk_cap en une variable cat√©gorique :

**1. Tr√®s Petite (Micro/Compacte) :**

Cargo capacity < 8 cubic feet

**2. Petite (Sous-compacte/Compacte) :**

8 <= Cargo capacity < 12 cubic feet

**3. Mod√©r√©e (Taille interm√©diaire) :**

12 <= Cargo capacity < 20 cubic feet

**4. Spacieuse (VUS de grande taille/Tailles interm√©diaires, Breaks) :**

20 <= Cargo capacity < 30 cubic feet

**5. Tr√®s Spacieuse (VUS de grande taille, Minifourgonnettes) :**

Cargo capacity >= 30 cubic feet 

```python
# D√©finir les intervalles et les √©tiquettes correspondantes
intervals = [float('-inf'), 8, 12, 20, 30, float('inf')]
labels = ['Very Small', 'Small', 'Moderate', 'Spacious', 'Very Spacious']

# Convertir la colonne 'trunk_cap' en une colonne cat√©gorique bas√©e sur les intervalles et les √©tiquettes
data_cleaned_3['trunk_cap_categ'] = pd.cut(data_cleaned_3['trunk_cap'], bins=intervals, labels=labels, right=False)
```


**FWD (Front-Wheel Drive) :** 

Dans un v√©hicule √† traction avant (FWD), la puissance du moteur est transmise aux roues avant. Cette configuration est courante dans de nombreuses voitures plus petites et offre une bonne traction dans la plupart des conditions.

**AWD (All-Wheel Drive) :**

La transmission int√©grale (AWD) signifie que la puissance est distribu√©e aux quatre roues du v√©hicule. Elle peut √™tre soit √† temps plein, o√π toutes les roues sont propuls√©es en permanence, soit √† temps partiel, o√π la puissance est envoy√©e aux roues qui en ont le plus besoin en fonction des conditions de la route.

**2WD (Two-Wheel Drive) :**

Ce terme est souvent utilis√© de mani√®re interchangeable avec 2RM (2 roues motrices), et il se r√©f√®re g√©n√©ralement aux v√©hicules √† traction avant (FWD) ou √† propulsion arri√®re (RWD).

**4WD (Four-Wheel Drive) :**

La transmission √† quatre roues motrices (4WD) signifie que les quatre roues sont propuls√©es, g√©n√©ralement utilis√©e dans les v√©hicules tout-terrain ou les camions pour une meilleure traction dans des terrains difficiles. Certains syst√®mes 4WD peuvent √™tre activ√©s ou d√©sactiv√©s en fonction des conditions.

**RWD (Rear-Wheel Drive) :**

Dans un v√©hicule √† propulsion arri√®re (RWD), la puissance du moteur est transmise aux roues arri√®re. Cette configuration est souvent utilis√©e dans les voitures de sport et les v√©hicules plus grands.

### **Engine**

Exemples :

**V6 (6-Cylinder) :**
Cela fait r√©f√©rence au nombre de cylindres dans le moteur. Dans ce cas, il s'agit d'un moteur √† six cylindres. Les moteurs √† six cylindres, comme les moteurs V6, sont connus pour leur puissance et leurs performances accrues par rapport aux moteurs √† quatre cylindres.

**4-Cyl (4-Cylinder) :** 
Cela fait r√©f√©rence au nombre de cylindres dans le moteur. Les moteurs peuvent avoir diff√©rents nombres de cylindres, et dans ce cas, il s'agit d'un moteur √† quatre cylindres.

```python
eng=[]
for c in data_cleaned_3.Engine :
    if ',' in c :
        eng.append(c.split(',')[0])
    elif '.' in c :
        eng.append(c.split('.')[0])
    elif '  '  in c :
        eng.append(c.split('  ')[0])
    else : 
        eng.append(c.split(' ')[0])
    
data_cleaned_3['Engine']=eng

data_cleaned_3['Engine']
```
